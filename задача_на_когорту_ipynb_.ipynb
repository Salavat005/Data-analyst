Допустим, мы работаем в интернет-магазине и хотим понять, сколько заказов и денег клиенты приносят в течение года после первой покупки. Для этого у нас есть данные о заказах:

```python
df['order_date'].max() # дата последнего заказа
```

```python
df['order_date'].min() # дата первого заказа
```

```python
df['order_date']= pd.to_datetime(df['order_date']) # преобразование формата даты
```

```python
df1=df.groupby(['customer_id'])['order_date'].min().to_frame().reset_index() # первые ордера клиентов
```

```python
df1
```

```python
df2=df.merge(df1, how='left', left_on='customer_id', right_on='customer_id') # добавляем в нашу таблицу столбец с датой первого заказа каждого клиента
```

```python
df2
```

```python
df2.groupby(['order_date_y','order_date_x'])['sales'].agg(['sum', 'count']) # группируем данные по первому заказу
```

```python
df2['кол-во дней'] = (df2['order_date_x'] - df2['order_date_y']).dt.days # добавляем столбец с количеством дней от первого заказа
```

```python
df2
```

```python
df3=df2[df2['кол-во дней']<=365] # убираем все,что больше 365 дней
```

```python
df3
```

```python
df4=df3.groupby(['order_date_y'])['sales'].agg(['sum', 'count']) 
```

```python
df4
```

```python
df4.groupby('order_date_y').sum().mean()
```

## Ответ: клиенты в течение года делают в среднем около 4 заказов и приносят 1949 уе денег))
